### 2) map()函数：新建一列

使用map()函数，由已有的列生成一个新列

适合处理某一单独的列。

```python
df = DataFrame(np.random.randint(0,150,size = (4,4)),columns=['Python','Java','PHP','HTML'],
               index = ['张三','旭日','阳刚','木兰'])
df
```

```python
#Go
#map也有映射关系，新添加一列，根据现存的那一列进行添加
v = {82:90,127:78,11:166,88:55}
df['Go'] = df['Python'].map(v)
```

```python
df
```

仍然是新建一个字典

map()函数中可以使用lambda函数

```python
#C
df['C'] = df['Go'].map(lambda x : x -40)
```

```python
df
```

```python
def mp(x):
    #复杂的条件
    if x <51:
        return '不及格'
    else:
        return '优秀'
```

```python
df['score'] = df['C'].map(mp)
df
```

transform()和map()类似

```python
df['score2'] = df['C'].transform(mp)
```

```python
df
```

使用map()函数新建一个新列

```python
# 同时map还可以修改当前列
# map()中不能使用sum之类的函数
df['C'] = df['C'].map(lambda x : x*2)
```

```python
df
```

## 5. 数据聚合【重点】

数据聚合是数据处理的最后一步，通常是要使每一个数组生成一个单一的数值。

数据分类处理：

 - 分组：先把数据分为几组
 - 用函数处理：为不同组的数据应用不同的函数以转换数据
 - 合并：把不同组得到的结果合并起来
 
数据分类处理的核心：
     groupby()函数

```python
df = DataFrame({'item':['苹果','香蕉','橘子','香蕉','橘子','苹果'], 'price':[4,3,3,2.5,4,2],'color':['red','yellow','yellow',
 'green','green','green']})
df
```

```python
g = df.groupby('item')
g.groups
```

```python
m = g['price'].mean()
type(m)
```

```python
df_mean = DataFrame(m)
df_mean.columns = ['price_mean']
df_mean
```

```python
pd.merge(df, df_mean, left_on='item', right_index=True)
```

如果想使用color列索引，计算price1的均值，可以先获取到price1列，然后再调用groupby函数，用参数指定color这一列

```python
df.groupby('color').sum()
```

使用.groups属性查看各行的分组情况：

```python
# 在进行分组的时候，如果多个条件进行分组，那么数据 就会有多层索引
df.groupby(['color', 'item']).sum()
```
