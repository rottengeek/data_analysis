查看日期格式，并将其转换为Pandas的日期格式,通过函数加map方式进行转换

```python
ele.dtypes
```

```python
def time_convert(dt):
    
    day,mon,year = dt.split('-')
    
    return '20'+year+'-'+str(months[mon])+'-'+day
```

```python
ele['contb_receipt_dt'] = ele['contb_receipt_dt'].map(time_convert)
```

查看是否转换成功
